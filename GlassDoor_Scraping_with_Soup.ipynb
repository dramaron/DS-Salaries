```python
soup = BeautifulSoup(html_text.content, 'lxml')
```

```python
company_name = []
for company in soup.find_all('div', class_='d-flex justify-content-between align-items-start'):
    company_name.append(company.text.strip())
company_name    
```

```python
job_title = []
for job in soup.find_all('a', class_='jobLink css-1rd3saf eigr9kq3'):
    job_title.append(job.text)
job_title    
```

```python
location = []
for loc in soup.find_all('span', 'css-3g3psg pr-xxsm css-iii9i8 e1rrn5ka0'):
    location.append(loc.text)
location    
```

```python
links = []
for link in soup.find_all('a', class_="jobLink css-1rd3saf eigr9kq3"):
    links.append('https://www.glassdoor.com' + link['href'])
links    
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.76 Safari/537.36'} # This is chrome, you can set whatever browser you like
#response = requests.get('http://www.rlsnet.ru/search_result.htm?word=%D6%E5%F0%E5%E1%F0%EE%EB%E8%E7%E8%ED', headers=headers)
description = []
for link in links:
    page = requests.get(link, headers=headers)
    #print(page.text)
    bs = BeautifulSoup(page.content, 'lxml')
    for job in bs.find_all('div', class_='desc css-58vpdc ecgq1xb5'):
        description.append(job.text)
description        
```

```python
for link in links:
    page = requests.get(link)
    print(page.status_code)
```

```python
salary = []
for link in links:
    page = requests.get(link, headers=headers)
    #print(page.text)
    bs = BeautifulSoup(page.content, 'lxml')
    for job in bs.find_all('span', class_='small css-10zcshf e1v3ed7e1'):
        salary.append(job.text.strip())
salary     
```
